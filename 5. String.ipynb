```python
# Whitespace
my_string = '   Hello World  '
print(my_string)
my_string = my_string.strip() # doesnot change in place. it make a copy or assign possible
print(my_string)
```

```python
my_string = "Hello World"
print(my_string.upper())
print(my_string.lower())
print(my_string.startswith("Hello")) # check it starts with
print(my_string.endswith("Hello")) # chek ends with
```

```python
# lets see list and string
my_string = 'how,are,you,doing'
my_list = my_string.split(",") #default is space
print(my_list)

new_string = ' '.join(my_list)
print(new_string)
```

```python
from timeit import default_timer as timer
my_list = ['a']*100000
print(my_list)

# bad way
start = timer()
my_string = ''
for i in my_list:
    my_string += i
stop = timer()
print(stop-start, my_string)

# good waymy_string
start = timer()
my_string = ''.join(my_list)
stop = timer()
print(stop-start, my_string)
```

```python
# Formatring string
# %, .format(), f-Strings
var = "Tom"
my_string = "the variable is %s" %var
print(my_string)
```

```python
var = 3
my_string = "the variable is %d" %var
print(my_string)
```

```python
var = 3.5
my_string = "the variable is %.2f" %var
print(my_string)
```

```python
my_string = "the variable is {}".format(var)
print(my_string)
```

```python
var2 = 6
my_string = "the variable is {:.2f} and {}".format(var, var2)
print(my_string)
```

```python
var2 = 6
my_string = f"the variable is {var*2} and {var2}"
print(my_string)
```
